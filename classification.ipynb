```python
import plotly.express as spx
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

from google.colab import drive
drive.mount('/content/drive')
```

```python
census_dataset = pd.read_csv("/content/drive/MyDrive/census.csv")
census_dataset
```

```python
X_census = census_dataset.iloc[:, 0:14].values
X_census
```

```python
y_census = census_dataset.iloc[:, 14].values
y_census
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()
```

```python
X_census[:, 1] = label_encoder_workclass.fit_transform(X_census[:, 1])
X_census[:, 3] = label_encoder_education.fit_transform(X_census[:, 3])
X_census[:, 5] = label_encoder_marital.fit_transform(X_census[:, 5])
X_census[:, 6] = label_encoder_occupation.fit_transform(X_census[:, 6])
X_census[:, 7] = label_encoder_relationship.fit_transform(X_census[:, 7])
X_census[:, 8] = label_encoder_race.fit_transform(X_census[:, 8])
X_census[:, 9] = label_encoder_sex.fit_transform(X_census[:, 9])
X_census[:, 13] = label_encoder_country.fit_transform(X_census[:, 13])
```

```python
X_census[0]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_census = scaler.fit_transform(X_census)
```

```python
X_census
```

```python
from sklearn.model_selection import train_test_split
X_census_train, X_census_test, y_census_train, y_census_test = train_test_split(X_census, y_census, test_size=0.15, random_state=0)
```

```python
X_census_train.shape, X_census_test.shape
```

### PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
X_census_train_pca = pca.fit_transform(X_census_train)
X_census_test_pca = pca.transform(X_census_test)
```

```python
X_census_train_pca.shape, X_census_test_pca.shape
```

```python
X_census_train_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
random_forest_census_pca = RandomForestClassifier(n_estimators=40, criterion='entropy', random_state=0)
random_forest_census_pca.fit(X_census_train_pca, y_census_train)
```

```python
predictions = random_forest_census_pca.predict(X_census_test_pca)
accuracy_score(y_census_test, predictions)
```

```python
predictions
```

```python
y_census_test
```

###Kernel PCA

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca = KernelPCA(n_components=8, kernel='rbf')
X_census_train_kpca = kpca.fit_transform(X_census_train)
X_census_test_kpca = kpca.transform(X_census_test)
```

###LDA
