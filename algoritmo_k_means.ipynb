```python
import plotly.express as spx
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.cluster import KMeans
```

```python
x = [20,27,21,37,46,53,55,47,52,32,39,41,39,48,48]
y = [1000, 1200, 1900, 1850, 900, 950, 2000, 2100, 3000,5000,4100,5100,7000,5000,6500]
```

```python
grapthc = spx.scatter(x=x,y=y)
grapthc.show()
```

```python
salaryDatabase = np.array([
    [20,1000],[27,1200],[21,2900],[37,1850],[46,900],[53,950],[55,2000],[47,2100],[52,3000],[32,5900],
    [39,4100],[41,5100],[39,7000],[48,5000],[48,6500],
])
```

```python
salaryDatabase
```

```python
salaryScaler = StandardScaler()
salaryDatabase = salaryScaler.fit_transform(salaryDatabase)
```

```python
salaryDatabase
```

```python
salaryKmeans = KMeans(n_clusters=3)
salaryKmeans.fit(salaryDatabase)
```

```python
centoides = salaryKmeans.cluster_centers_
```

```python
centoides
```

```python
salaryScaler.inverse_transform(salaryKmeans.cluster_centers_)
```

```python
labels = salaryKmeans.labels_
```

```python
labels
```

```python
grapthc_1 = spx.scatter(x=salaryDatabase[:,0],y=salaryDatabase[:,1],color=labels)
grapthc_2 = spx.scatter(x=centoides[:,0],y=centoides[:,1], size=[12, 12, 12])
#grapthc_1.show()
grapthc_3 = go.Figure(data=grapthc_1.data + grapthc_2.data)
grapthc_3.show()
```

```python
creadCardDatabase = pd.read_csv('/content/drive/MyDrive/credit_card_clients.csv', header=1)
creadCardDatabase
```

```python
creadCardDatabase['BILL_TOTAL'] = creadCardDatabase['BILL_AMT1'] + creadCardDatabase['BILL_AMT2'] + creadCardDatabase['BILL_AMT3'] + creadCardDatabase['BILL_AMT4'] + creadCardDatabase['BILL_AMT5'] + creadCardDatabase['BILL_AMT6']
```

```python
creadCardDatabase
```

```python
X_credCard = creadCardDatabase.iloc[:,[1,25]].values
X_credCard
```

```python
credCarScaler = StandardScaler()
X_credCard = credCarScaler.fit_transform(X_credCard)
X_credCard
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(X_credCard)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
grapthc = spx.line(x=range(1,11), y=wcss)
grapthc.show()
```

```python
credCardKeyMeans = KMeans(n_clusters=4, random_state=0)
credCardKeyMeans.fit_predict(X_credCard)
```

```python
grapthc = spx.scatter(x=X_credCard[:,0],y=X_credCard[:,1],color=credCardKeyMeans.labels_)
grapthc.show()
```

```python
clientList = np.column_stack((creadCardDatabase,credCardKeyMeans.labels_))
clientList
```

```python
clientList = clientList[clientList[:, 26].argsort()]
clientList
```

```python
from sklearn.datasets import make_blobs
```

```python
X_random, y_random = make_blobs(n_samples=200, centers=5, random_state=1)
```

```python
X_random
```

```python
y_random
```

```python
graphic = spx.scatter(x=X_random[:,0], y=X_random[:,1])
graphic.show()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X_random)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
labels = kmeans.labels_
```

```python
graphic1 = spx.scatter(x=X_random[:,0], y=X_random[:,1], color=labels)
graphic2 = spx.scatter(x=centroids[:,0], y=centroids[:,1], size=[12, 12, 12, 12, 12])
graphic3 = go.Figure(data=graphic1.data + graphic2.data)
graphic3.show()
```
